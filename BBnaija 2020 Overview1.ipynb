```python
import wordcloud
from wordcloud import WordCloud,STOPWORDS
from PIL import Image
```

### Ozo's Wordcloud

```python
new_df = df[(df['username']=='OfficialOzoBBN')]
```

```python
def color_func(word=None, font_size=None,position=None,orientation=None, font_path=None,random_state=None):
    h = int(240.0 * 255.0 / 255.0)
    s = int(100.0 * 255.0 / 255.0)
    #l = int(100.0 * float(random_state.randint(60, 120)) / 250)
    l = int(50.0 * 255.0 / 255.0)
    return 'hsl({}, {}%, {}%)'.format(h, s, l)
```

```python
with open('ozo_wc.txt','w') as f:
    for row in new_df.text:
        f.write(row)
    f.close()
```

```python
ozo_text = open('ozo_wc.txt', 'r').read()
stopwords = set(STOPWORDS)
more_stopwords = ['pic', 'twitter', 'com']
for word in more_stopwords:
    stopwords.add(word)
```

```python

ozo_wc = WordCloud(background_color='white', stopwords=stopwords, 
                   color_func=color_func)
ozo_wc.generate(ozo_text)
#display the word cloud
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(ozo_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
